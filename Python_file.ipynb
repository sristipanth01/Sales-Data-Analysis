Data visualization

A box plot (also known as a whisker plot) provides a visual summary of the distribution of data and highlights potential outliers. The key components of a box plot includes Box,Whiskers and outliers.

```python
import pandas as pd
sales=pd.read_csv(r'C:\Users\Dell\Desktop\sales_dataset.csv')
Q1 = sales['Sales'].quantile(0.25)  
Q3 = sales['Sales'].quantile(0.75)  
IQR = Q3 - Q1  # Interquartile range

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = sales[sales['Sales'] > upper_bound]
print(outliers)
```

```python
sales['Sales'].plot(kind='box',vert=False,figsize=(14,6))
```

From above Visulization we get:
1. Most sales values are concentrated at the lower range, as indicated by the narrow box and dense clustering near the median.
2. A significant number of outliers exist, representing high sales values far from the main data distribution (to the right). This suggests occasional large transactions or orders.
3. The distribution of sales is positively skewed, as the whisker extends much further to the right, showing a long tail of higher sales values.
4. The compact interquartile range (IQR) indicates consistent sales for most transactions, except for a few large-value outliers.

Density Plot: A density plot is a graphical representation that shows the distribution of a continuous variable. It estimates the probability density function of the variable, providing insights into its underlying distribution.

```python
sales['Sales'].plot(kind='density', figsize=(14, 6))
```

1. Most sales are concentrated around low values, indicating frequent smaller purchases.
2. A long right tail suggests rare but significant high-value transactions.
3. Negative values might represent refunds, returns, or data errors requiring further investigation.
4. Positively skewed data indicates most sales occur at lower price points.
5. Small transactions dominate in frequency, but high-value sales contribute significantly to revenue.
6. Potential to introduce mid-tier products or adjust pricing strategies to capture more sales.

Ensuring proper date formats

```python
import pandas as pd
import numpy as np


sales=pd.read_csv(r'C:\Users\Dell\Desktop\sales_dataset.csv')
sales['Order_Date'] = pd.to_datetime(sales['Order_Date'])
sales.head()
```

Extract daily, weekly, or monthly periods from the Order_Date column

```python
# Daily trends
daily_sales = sales.groupby('Day')['Sales'].sum()

# Weekly trends
weekly_sales = sales.groupby('Week')['Sales'].sum()

# Monthly trends
monthly_sales = sales.groupby('Month')['Sales'].sum()
sales.head()
```

Daily Sales Trends

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(daily_sales.index, daily_sales.values, marker='o', label='Daily Sales')
plt.title('Daily Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid()
plt.legend()
plt.show()
```

This line chart visualizes the trend of total daily sales over time. Each point on the chart represents the total sales for a specific day. 

- The x-axis displays the dates.
- The y-axis shows the total sales for each day.
- The chart includes markers for each data point, a grid for better readability, and a legend for clarity.

```python
monthly_sales_change = monthly_sales.pct_change() * 100
print(monthly_sales_change)
```

This code calculates the month-over-month percentage change in total sales. The calculation helps identify the growth or decline rate in sales between consecutive months.

- **`pct_change()`**: Computes the percentage change between the current and previous elements in the `monthly_sales` series.
- **`* 100`**: Converts the ratio to a percentage.

The result is printed as a series where:
- The index represents the months.
- The values represent the percentage change in sales compared to the previous month.

Comparing customer purchasing patterns. 

```python
sales_cleaned = sales.dropna(subset=['Customer_Name', 'Order_Date', 'Sales'])
sales_cleaned['Order_Date'] = pd.to_datetime(sales_cleaned['Order_Date'])
```

Total spend by customers

```python
total_spending = sales_cleaned.groupby('Customer_Name')['Sales'].sum().sort_values(ascending=False)
print("Top 5 Customers by Total Spending:")
print(total_spending.head())
```

Purchase Frequency

```python
purchase_frequency = sales_cleaned.groupby('Customer_Name')['Order_ID'].nunique().sort_values(ascending=False)
print("Top 5 Customers by Purchase Frequency:")
print(purchase_frequency.head())
```

Average Order Value

```python
average_order_value = total_spending / purchase_frequency

customer_patterns = pd.DataFrame({
    'Total Spending': total_spending,
    'Purchase Frequency': purchase_frequency,
    'Average Order Value': average_order_value
})

# Display the top 5 customers
print(customer_patterns.head())
```

 Compare Patterns Across Customer Segments

```python
high_spenders = customer_patterns[customer_patterns['Total Spending'] > 1000]
frequent_buyers = customer_patterns[customer_patterns['Purchase Frequency'] > 10]
low_spenders = customer_patterns[customer_patterns['Total Spending'] < 100]

# Analyze each segment
print(f"Number of High Spenders: {len(high_spenders)}")
print(f"Number of Frequent Buyers: {len(frequent_buyers)}")
print(f"Number of Low Spenders: {len(low_spenders)}")
```

How do product categories contribute to sales? 

```python
category_sales = sales.groupby('Product_Category')['Sales'].sum()
category_sales_percentage = (category_sales / category_sales.sum()) * 100
category_sales_sorted = category_sales.sort_values(ascending=False)

print(category_sales_sorted)
print(category_sales_percentage)
```

```python
# Visualization: Bar Plot
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
category_sales_sorted.plot(kind='bar', color='skyblue')
plt.title('Contribution of Product Categories to Sales')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

The Bar chart above represents the contribution of Product Categories to the sales.
- The x-axis represents the Product Category.
- The y-axis represents the Total Sales.

```python
# Visualization: Pie Chart
plt.figure(figsize=(8, 8))
category_sales_percentage.plot(kind='pie', autopct='%1.1f%%', startangle=140, colormap='coolwarm')
plt.title('Percentage Contribution of Product Categories to Sales')
plt.ylabel('')  
plt.show()
```

The chart above visualizes the contribution of Product Categories to Sales. It uses the category_sales_percentage function to convert the sales value to percentage.The plot(kind='pie') function then creates a pie chart, with the percentage share of each product categories.

Pie-chart for customer segmentation

```python
import matplotlib.pyplot as plt
import pandas as pd

sales['Segment'] = pd.cut(
    sales['Sales'],
    bins=[0, 200, 500, 1000],
    labels=['Low', 'Medium', 'High']
)

segment_counts = sales['Segment'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(
    segment_counts, 
    labels=segment_counts.index, 
    autopct='%.1f%%', 
    colors=['#FFD700', '#87CEEB', '#FF6347']
)
plt.title('Customer Segmentation', fontsize=16)
plt.show()
```

This pie chart visualizes the segmentation of customers based on their total sales amounts. Customers are categorized into three segments:

- **Low**: Sales below $200
- **Medium**: Sales between $200 and $500
- **High**: Sales above $500

- **`pd.cut()`**: Categorizes sales values into the defined segments.
- **`value_counts()`**: Counts the number of customers in each segment.

The chart uses three distinct colors to represent the segments, with percentages displayed for each.
